```python
import h5py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.offsetbox import TextArea, DrawingArea, OffsetImage, AnnotationBbox
from matplotlib.offsetbox import AnchoredText
from tqdm import tqdm
import matplotlib.image as mpimg
import warnings
import pathlib
path = pathlib.Path.cwd()

plt.rcParams['font.family'] = 'serif'
plt.rcParams['figure.autolayout'] = True
plt.rcParams.update({'font.size': 14})
np.seterr(divide = 'ignore') 
warnings.filterwarnings('ignore')
```

```python
def plot_EFPs(bbx, N=None):
    hf_file = path.parent / "data" / "EFP" / f"EFP-{bbx}.h5"
    if hf_file.exists():
        hf = h5py.File(hf_file, "r")
        threshold = 0.001
        nbins = 30
        colors = {0:"r", 1:"b", 2:"k"}
        y = hf["targets"][:N]
        targets = list(set(y))
        efps = hf["efp"]
        for ix, efp in enumerate(tqdm(efps.keys())):
            fig, ax = plt.subplots(figsize=(6,5))
            X = np.log10(efps[efp][:N])
            X[X == -np.inf] = -100
            b0, h0 = np.histogram(X, nbins, density=True)
            h0 = h0[:-1][b0 > threshold]
            bins = np.linspace(min(h0), max(h0), nbins)
            df = pd.DataFrame({"efp":X, "y":y})
            grouped = df.groupby(["y"])        
            for target in targets:
                X = grouped.get_group(target).efp.values
                histtype = "step"
                if target == 0:
                    histtype = "stepfilled"
                ax.hist(X, bins=bins, density=True, histtype=histtype, linewidth=3, label=f"y={target}", color=colors[target])
            ax.set_ylabel("Density")
            ax.set_xlabel("log$_{10}$(EFP)")
            ymax = ax.get_ylim()[-1]
            ax.set_ylim(top=ymax*1.5)
            ax.legend(frameon=False)
            arr_lena = mpimg.imread(path.parent / "graphs" / "png" / f"efp_{efp[:5]}.png")
            kb = efp.split("_")
            kappa = kb[-3]
            beta = kb[-1]
            imagebox = OffsetImage(arr_lena, zoom=0.5)
            xlim = ax.get_xlim()
            xpos = xlim[0] + np.absolute(xlim[0] - xlim[1])*0.1
            ylim = ax.get_ylim()
            ypos = ylim[-1]*0.86
            ab = AnnotationBbox(imagebox, (xpos, ypos))
            kb_text = "$\\kappa=%s$\n$\\beta=%s$" %(kappa, beta)
            at = AnchoredText(kb_text, loc="lower left", frameon=False, prop=dict(size=16), bbox_to_anchor=(0.16, 0.72), bbox_transform=ax.transAxes)
            ax.add_artist(at)
            ax.add_artist(ab)
            plt.draw()
            plt.show()
            plt.savefig(path.parent / "figures" / "EFP" / bbx / "png" / f"{efp}.png")
            plt.savefig(path.parent / "figures" / "EFP" / bbx / "pdf" / f"{efp}.pdf")
            plt.clf()
        hf.close()
        
for bbx in ("bb1", "bb3"):
    plot_EFPs(bbx)
```
