```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
import pathlib
from matplotlib.colors import LogNorm
from mpl_toolkits.axes_grid1 import make_axes_locatable
path = pathlib.Path.cwd()
plt.rcParams['font.family'] = 'serif'
plt.rcParams['figure.autolayout'] = True
plt.rcParams.update({'font.size': 12})
```

```python
hf = h5py.File("jet_images_chan_1.h5", "r")
N = 50000
cmap = "gist_heat"
for key in hf.keys():
    j0 = np.zeros((32,32,1))
    b0 = np.zeros((32,32,1))
    jet_images = hf[key]["features"][:N]
    targets = hf[key]["targets"][:N]
    for X, y in zip(jet_images, targets):
        if y == 0:
            b0+= X
        elif y == 1:
            j0 += X
    fig, ax = plt.subplots(1, 2, figsize=(8,4), sharex=True, sharey=True)
    b0 /= np.sum(b0)
    j0 /= np.sum(j0)
    vmin = min(np.min(b0), np.min(j0))
    vmax = max(np.max(b0), np.max(j0))
    bkg = ax[0].imshow(b0, norm=LogNorm(vmin, vmax), cmap=cmap)
    sig = ax[1].imshow(j0, norm=LogNorm(vmin, vmax), cmap=cmap)
    fig.colorbar(sig, cax=cax, label="$E_T$")
    fig.subplots_adjust(right=0.8)
    plt.suptitle("$r_{inv} = %s$" %key.replace("p","."))
    ax[0].set_title("Bkg.")
    ax[1].set_title("Sig.")
    ax[0].set_yticklabels([])
    ax[0].set_xticklabels([])
    ax[1].set_yticklabels([])
    ax[1].set_xticklabels([])
    #plt.tight_layout()
    #plt.subplots_adjust(wspace=0.0)
    plt.savefig(path.parent / "figures" / "jet_images" / f"{key}.png")
    plt.savefig(path.parent / "figures" / "jet_images" / f"{key}.pdf")
hf.close()
```

```python
hf = h5py.File("jet_images_chan_2.h5", "r")
for N in (2, 20000):
    cmap = "gist_heat"
    for key in hf.keys():
        j0 = np.zeros((32,32))
        b0 = np.zeros((32,32))
        j1 = np.zeros((32,32))
        b1 = np.zeros((32,32))
        jet_images = hf[key]["features"][:N]
        targets = hf[key]["targets"][:N]
        for X, y in zip(jet_images, targets):
            if y == 0:
                b0+= X[:,:,0]
                b1+= X[:,:,1]
            elif y == 1:
                j0 += X[:,:,0]
                j1 += X[:,:,1]
        fig, ax = plt.subplots(2, 2, figsize=(8,8), sharex=True, sharey=True)
        vmin = min(np.min(b0), np.min(j0))
        vmax = max(np.max(b0), np.max(j0))
        ax[0,0].imshow(b0, norm=LogNorm(), cmap=cmap)
        ax[0,1].imshow(j0, norm=LogNorm(), cmap=cmap)
        ax[1,0].imshow(b1, norm=LogNorm(), cmap=cmap)
        ax[1,1].imshow(j1, norm=LogNorm(), cmap=cmap)
        fig.subplots_adjust(right=0.8)
        plt.suptitle("$r_{inv} = %s$  N=%s" %(key.replace("p","."), N))
        ax[0,0].set_title("Chan 1 - Bkg")
        ax[0,1].set_title("Chan 1 - Sig")
        ax[1,0].set_title("Chan 2 - Bkg")
        ax[1,1].set_title("Chan 2 - Sig")
        ax[0,0].set_yticklabels([])
        ax[0,0].set_xticklabels([])
        ax[0,1].set_yticklabels([])
        ax[0,1].set_xticklabels([])
        ax[1,0].set_yticklabels([])
        ax[1,0].set_xticklabels([])
        ax[1,1].set_yticklabels([])
        ax[1,1].set_xticklabels([])
        plt.savefig(path.parent / "figures" / "jet_images" / f"2ch_{key}_n_{N}.png")
        plt.savefig(path.parent / "figures" / "jet_images" / f"2ch_{key}_n_{N}.pdf")
hf.close()
```
